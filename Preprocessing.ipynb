```python
import pandas as pd
```

```python
nike = pd.read_csv('nike_shoes_dataset.csv', sep=',')
nike
```

```python
nike = nike.rename(columns={'Product ID':'product_id', 'Product Name':'product_name', 'Category':'category',
                           'Price' : 'MRP', 'Discount': 'discount', 'Final Price': 'selling_price', 
                           'Release Date' : 'release', 'Shoe Size' : 'size', 'Color' : 'color',
                           'Material' : 'material', 'Gender' : 'gender', 'Rating' : 'rating', 
                           'Stock':'stock', 'Sold Quantity' : 'quantity'})
```

```python
import numpy as np

nike['selling_price'] = np.where(
    nike['discount'] > 50, 
    nike['MRP'] * (nike['discount'] / 100), 
    nike['MRP'] * ((100 - nike['discount']) / 100)
)
```

```python
nike['discount'] = np.where(
    nike['discount'] > 50, 
    100 - nike['discount'], 
    nike['discount']
)
```

```python
nike['release'] = pd.to_datetime(nike['release'])
```

```python
nike['release'] = nike['release'].dt.strftime('%d-%m-%Y')
```

```python
nike.to_csv('nike.csv')
```
